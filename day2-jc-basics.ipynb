```python
import pandas as pd

# Sample ERP-style transaction data
data = {
    "OrderID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Customer": ["A", "B", "C", "A", "B", "E", "A", "B", "B", "C"],
    "Amount": [250, 450, 300, 150, 20, 40, 60, 200, 100, 520],
    "Status": ["Shipped", "Pending", "Shipped", "Cancelled", "Open", "Shipped", "Pending", "Shipped", "Shipped", "Null"]
}

df = pd.DataFrame(data)

print("üîç Original Data")
print(df)
print()   # Give blank line 

print("üîΩ Sorted by Amount (High to Low)")
df_sorted = df.sort_values("Amount", ascending=False)
print(df_sorted)
print()   # Give blank line 

print("üîç Filtered: Status = 'Pending'")
df_pending = df[df["Status"] == "Pending"]
print(df_pending)
print()   # Give blank line 

print("üìä Grouped: Total Amount by Customer")
df_grouped = df.groupby("Customer")[["Amount"]].sum().reset_index()
print(df_grouped)
print()   # Give blank line 

df["Discount"] = df["Amount"] * 0.3
print("üí∏ Data with Discount Column")
print(df)
print()   # Give blank line 

print("üìä Grouped: Total Discount by Customer")
df_discount_summary = df.groupby("Customer")[["Discount"]].sum().reset_index()
print(df_discount_summary)
```
